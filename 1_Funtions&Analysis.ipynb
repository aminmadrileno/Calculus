## Welcome to Functions & Analysis in Python! 
#### We will start by taking a look at linear equations, finding the x-intercept and how to plot them. 

```python
import numpy as np
import matplotlib.pyplot as plt

# first I am defining the linear function that will compute f(x) = m*x + b.
def linear_function(x, m, b):
    return m * x + b

# I set the parameters for the linear function.
m = 1.5    # That will be the slope of the graph.
b = 2   # and my y-intercept.

# Now I am generating a range of x values.
# np.linspace is basically just generating 300 values between -5 and 5 equally distributedd... so i will have 300 x-values
x_values = np.linspace(-5, 5, 300)

# I compute the corresponding y values for each x using the linear function with respect to m,b...  so basically im running the function.
y_values = linear_function(x_values, m, b)

# Next, I calculate the x-axis-interception).
# For a linear function, the x-intercept is where m*x + b = 0,
# so I solve for x which gives me x = -b/m (assuming m is not zero).
x_intersect = -b / m
y_intersect = 0  # I know that at the x-axis, y is always zero.

# Now I create the plot and set its size.
plt.figure(figsize=(10, 6))

# I plot the linear function using the computed x and y values and stuff like color and label and so on.
plt.plot(x_values, y_values, label=f'f(x) = {m}x + {b}', color='blue')

# I highlight the x-intercept by adding a scatter point at (x_intersect, 0) so that it will clearly show the x-intercept.
# i set parameterslike color and zorder which is basically just saying that the point will be set above other lines. x-intercept with 2 decimal places
plt.scatter(x_intersect, y_intersect, color='red', zorder=5, 
            label=f'X-intercept: ({x_intersect:.2f}, 0)')

# I add horizontal and vertical lines to represent the x and y axes, make it green so it stands out.
plt.axhline(0, color='green', linewidth=1.5)
plt.axvline(0, color='green', linewidth=1.5)

# I set the labels for the x-axis and y-axis, and I also add a title to the plot.
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of the Linear Function')

# I add a legend and a grid to improve the readability of the plot.
plt.legend()
plt.grid(True)
plt.show()
```

### lets continue now with exponential equations

```python
import sympy as sp

# First, I again define a symbolic variable x using sympy.
x = sp.symbols('x')

# Now I define my exponential function: f(x) = 2 * 3^x.
# Here, 'a' is 2 and 'b' is 3.
a = 2
b = 3
f = a * b**x  # This should  represents 2 * 3^x.

# Next, I compute the derivative of the exponential function with respect to x. Here i also just use the .diff() function of sympy
f_prime = sp.diff(f, x)
print("The derivative of f(x) = 2 * 3^x is:", f_prime)

# I then convert the symbolic expressions to numerical functions using lambdify.
# This allows me to evaluate them over an array of x values.
f_lambda = sp.lambdify(x, f, 'numpy')
f_prime_lambda = sp.lambdify(x, f_prime, 'numpy')

# Now I generate a range of x values for plotting.
# I choose an interval from -5 to 5
x_values = np.linspace(-5, 5, 300)

# I compute the corresponding y values for the exponential function and its derivative.
y_values = f_lambda(x_values)
y_prime_values = f_prime_lambda(x_values)

# I create a plot to visualize both the exponential function and its derivative.
plt.figure(figsize=(10, 6))

# I plot the exponential function f(x).
plt.plot(x_values, y_values, label=f'f(x) = {a} * {b}^x', color='blue')

# I plot the derivative f'(x) as a dashed red line.
plt.plot(x_values, y_prime_values, label=f"f'(x)={sp.pretty(f_prime)}", color='red', linestyle='--')

# I add horizontal and vertical lines to represent the axes.
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# I add labels for the x and y axes, a title, and a legend to explain the plot.
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential Function and Derivative')
plt.legend()
plt.grid(True)

# Finally, I display the plot.
plt.show()
```

# Integral Calculus! Now we calculate an integral of a quadratic equation

```python
import sympy as sp

# First, I again define the symbolic variable using sympy.
x = sp.symbols('x')

# Now I define my function: f(x) = x^2 - 4x + 3.
f = x**2 - 4*x + 3

# I compute the indefinite integral (antiderivative) of f(x). that works using the integrate function of sympy
F = sp.integrate(f, x)
print("The indefinite integral (antiderivative) of f(x) is:" + str(F))

# Next, I compute the definite integral (area under the curve) of f(x)
# between x = 0 and x = 5.
definite_integral = sp.integrate(f, (x, 0, 5))
print("\nThe definite integral of f(x) from 0 to 5 is:", definite_integral)

# I now convert both the function and its antiderivative to numerical functions
# using lambdify, which allows me to evaluate them over an array of x values.
f_lambda = sp.lambdify(x, f, 'numpy')
F_lambda = sp.lambdify(x, F, 'numpy')

# I generate a range of x values for plotting.
x_values = np.linspace(-1, 6, 400)
y_values = f_lambda(x_values)

# I create the plot to visualize the function.
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label="f(x) = x^2 - 4x + 3", color='blue')

# Now I shade the area under the curve between x = 0 and x = 5.
# I generate x values within the interval [0, 5] for the shading.
ix = np.linspace(0, 5, 400)
iy = f_lambda(ix)
plt.fill_between(ix, iy, color='orange', alpha=0.3, label="Area from x=0 to x=5")

# I add horizontal and vertical lines for the axes.
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# I label my axes, add a title, and include a legend.
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Integral Calculus: Function and Area Under the Curve")
plt.legend()
plt.grid(True)

# Finally, I display the plot.
plt.show()
```
