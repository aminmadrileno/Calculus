```python
import numpy as np
import matplotlib.pyplot as plt

def linear_function(m, b, x):
    return m * x + b

# Generate x values
x = np.linspace(0, 15, 100)
y = linear_function(2, 6, x)  # Example: f(x) = 2x + 3

# Plot
plt.plot(x, y, label='f(x) = 2x + 3', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Linear Function')
plt.legend()
plt.grid()
plt.show()
```

```python
def quadratic_function(a, b, c, x):
    return a * x**2 + b * x + c

x = np.linspace(-10, 10, 100)
y = quadratic_function(1, -3, -4, x)  # Example: f(x) = x^2 - 3x - 4

plt.plot(x, y, label='f(x) = x^2 - 3x - 4', color='red')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Quadratic Function')
plt.legend()
plt.grid()
plt.show()
```

```python
def exponential_function(a, b, x):
    return a * b**x

x = np.linspace(-2, 2, 100)
y = exponential_function(2, 3, x)  # Example: f(x) = 2 * 3^x

plt.plot(x, y, label='f(x) = 2 * 3^x', color='green')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Exponential Function')
plt.legend()
plt.grid()
plt.show()
```
